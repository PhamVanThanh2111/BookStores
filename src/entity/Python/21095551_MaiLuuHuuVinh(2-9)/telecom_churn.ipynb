```python
d = {"No": False, "Yes": True}
df["International plan"] = df["International plan"].map(d)
df.head()
# Đoạn mã d = {"No": False, "Yes": True} được sử dụng để tạo ra một từ điển có tên là d với các cặp key-value là "No": False và "Yes": True. 
# Sau đó, phương thức df["International plan"] = df["International plan"].map(d) được sử dụng để thay thế các giá trị
# trong cột "International plan" của DataFrame df bằng các giá trị tương ứng trong từ điển d. Phương thức này có thể được sử dụng để 
# thay thế các giá trị trong DataFrame bằng các giá trị tương ứng trong từ điển.
```

```python
df = df.replace({"Voice mail plan": d})
df.head()
# Đoạn mã df = df.replace({"Voice mail plan": d}) được sử dụng để thay thế các giá trị trong cột "Voice mail plan" của
# DataFrame df bằng các giá trị tương ứng trong từ điển d.
# Phương thức này có thể được sử dụng để thay thế các giá trị trong DataFrame bằng các giá trị tương ứng trong từ điển.
```

```python
total_calls = (
    df["Total day calls"]
    + df["Total eve calls"]
    + df["Total night calls"]
    + df["Total intl calls"]
)
df.insert(loc=len(df.columns), column="Total calls", value=total_calls)
# loc parameter is the number of columns after which to insert the Series object
# we set it to len(df.columns) to paste it at the very end of the dataframe
df.head()
# Phương thức df.insert(loc=len(df.columns), column="Total calls", value=total_calls) 
# được sử dụng để thêm cột "Total calls" vào cuối DataFrame df. Phương thức này có thể được sử dụng để thêm cột vào DataFrame.
```

```python
df["Total charge"] = (
    df["Total day charge"]
    + df["Total eve charge"]
    + df["Total night charge"]
    + df["Total intl charge"]
)
df.head()
# Phương thức df["Total charge"] = (df["Total day charge"] + df["Total eve charge"] + 
# df["Total night charge"] + df["Total intl charge"]) được sử dụng để thêm cột "Total charge" vào cuối DataFrame df. 
# Phương thức này có thể được sử dụng để thêm cột vào DataFrame.
```
